## Start with C18 Neg ionization

Treat two columns as two libraries

```python
import pandas as pd

# use Ken's C18 neg library, not perfectly formatted
confirmed = 'IROA-Ken-2017/C18Neg_Orbitrap_IROA_Ken20170623.txt'

# check encoding
with open(confirmed) as f:
    print(f)
```

```python
# use engine='python' to deal with encoding issue
mytable = pd.read_table(confirmed, engine='python')
print(mytable.columns)
mytable.head()
# data row 1: 636
```

```python
mytable.tail(10)
```

```python
# wanted cols (3, 7, 8, 9, 10, 12, 19, 20)
# CNAME, M-H	Retention Time (C18)	PARENT_CID, HMDB, expected_conc, Preferred Adduct	m/z

d = mytable.iloc[0,[3, 7, 8, 9, 10, 12, 19, 20]]
print(d)
print(list(d))
```

```python
cpds = []
for ii in range(635):
    # CNAME, M-H, Retention Time (C18), PARENT_CID,  HMDB, expected_conc, Preferred Adduct, m/z
    d = list( mytable.iloc[ii, [3, 7, 8, 9, 10, 12, 19, 20]] )
    
    #if type(d[6]) != str:
    #    print (d)
    d = [str(x) for x in d]
    
    if d[2] != 'nan' and d[2].strip() != "*":  # control for observed rtime
        # if alternative ion is preferred, it precedes default M-H, per source doc
        # 
        
        # pandas takes nan for missing values
        
        if d[6] != 'nan' and d[6].strip():  # control for Preferred Adduct; 
            cpds.append( {
                'name': d[0],
                'observed_ion': d[6],
                'observed_rtime': d[2],
                'parent_KEGG': d[3],
                'HMDB': d[4],
                'HMDB_conc_range (uM)': d[5],
                'observed_mz': d[7],  # take m/z from preferred ion
            } )
        else:
            cpds.append( {
                'name': d[0],
                'observed_ion': 'M-H',  # default for neg ionization
                'observed_rtime': d[2],
                'parent_KEGG': d[3],
                'HMDB': d[4],
                'HMDB_conc_range (uM)': d[5],
                'observed_mz': d[1],  # take m/z from M-H column
            } )

print("Got %d confirmed metabolites in library." %len(cpds))
print(cpds[33])
```

```python
C18_neg = {
    'metadata': metadata,
    'compounds': cpds,
}
```
